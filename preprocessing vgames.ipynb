```python
with open('Downloads/vgsales.csv', 'r') as source_file:
    lines = source_file.readlines()
```

```python
commas = 0
cnt_lines = 0
crushed = []
for i in range(len(lines)):
    commas=lines[i].count(',')
    if commas!=10:
        cnt_lines+=1
        print(commas, lines[i])
        crushed.append(i)
cnt_lines
```

```python
for i in range(len(lines)):
    if i in crushed:
        
        temp0=lines[i].replace('40,000','40000')
        temp0=temp0.replace('Seyo,Otome','Seyo, Otome')
        temp0=temp0.replace('Hamsters,big','Hamsters, big')
        temp1=temp0.replace(', ','^')
        temp2=temp1.replace(',','|')
        temp3=temp2.replace('^',', ')
        lines[i]=temp3
    else:
        lines[i]=lines[i].replace(',','|')
```
